```python
import pandas as pd
import os
import glob


PROGRAM_DIRECTORY = os.getcwd()
os.chdir(PROGRAM_DIRECTORY)
```

```python

neutral_day_combined_hourly_data = pd.read_csv(f"{PROGRAM_DIRECTORY}\\ATC_2023_Combined.csv")
neutral_day_combined_hourly_data["site-Source"] = neutral_day_combined_hourly_data["site"].astype(str) + " - " + neutral_day_combined_hourly_data["Source"]

filtered_data_combined = pd.read_csv(f"{PROGRAM_DIRECTORY}\\ATC_2023_Combined_filtered_outliers_database.csv")
filtered_data_combined["site-Source"] = filtered_data_combined["site"].astype(str) + " - " + filtered_data_combined["Source"]

unfiltered_data_15m = pd.read_csv(f"{PROGRAM_DIRECTORY}\\ATC_2023_15m_appended_neutral_mon-thurs.csv")
unfiltered_data_15m["site-Source"] = unfiltered_data_15m["site"].astype(str) + " - " + unfiltered_data_15m["Source"]

filtered_data_15m = pd.read_csv(f"{PROGRAM_DIRECTORY}\\ATC_2023_15m_appended_neutral_mon-thurs_filtered_outliers_database.csv")
filtered_data_15m["site-Source"] = filtered_data_15m["site"].astype(str) + " - " + filtered_data_15m["Source"]

display(filtered_data_15m["site-Source"])

site_filter = pd.read_csv(f"{PROGRAM_DIRECTORY}\\util\\A10 2023 site filter\\A10_Microsim_Site_Filter.csv")

site_filter = site_filter[["site-Source"]]
display(site_filter)
```

```python
A10_combined_unfiltered_data = pd.merge(neutral_day_combined_hourly_data, site_filter, on = "site-Source", how = "inner")

A10_combined_filtered_data = pd.merge(filtered_data_combined, site_filter, on = "site-Source", how = "inner")

A10_15m_filtered_data = pd.merge(filtered_data_15m, site_filter, on = "site-Source", how = "inner")

A10_15m_unfiltered_data = pd.merge(unfiltered_data_15m, site_filter, how = "inner", on = "site-Source")

#A10_non_avg_data = pd.merge(non_avg_data, site_filter, on = ["site", "Source", "site-Source"], how = "inner")

#A10_non_avg_data.to_csv("A10_study_Area_Outlier_Filtered_Raw_Data.csv", index = False)

```
