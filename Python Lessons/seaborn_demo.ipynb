```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
database = pd.read_csv('ATC_1.csv', index_col=0, encoding='latin').drop(columns=["Name"])
database
```

```python
sns.set_style('whitegrid')
```

```python
car = database["Car"]
lgv = database["LGV"]
plt.scatter(car,lgv)
plt.xlabel("Car")
plt.ylabel("LGV")
plt.show
```

```python
sns.lmplot(x='Car', y='LGV', data=database)
```

```python
sns.lmplot(x='Car', y='LGV', data=database, fit_reg=False, hue='Stage')
```

```python
sns.lmplot(x='Car', y='LGV', data=database, fit_reg=False, hue='Stage')
plt.ylim(0,200)
plt.xlim(0,160)
```

```python
sns.boxplot(data=database)
```

```python
stats_database = database.drop(columns=["Total", "Stage", "Daily?"])
sns.boxplot(data=stats_database)
```

```python
sns.set_style('darkgrid')
sns.boxplot(data=stats_database)
```

```python
road_key_colors = ['#78C850',  
                    '#F08030',  
                    '#6890F0',  
                    '#A8B820',  
                    '#A8A878',  
                    '#A040A0',  
                    '#F8D030',  
                    '#E0C068',  
                    '#EE99AC',  
                    '#C03028',  
                    '#F85888',  
                    '#B8A038',  
                    '#705898',  
                    '#98D8D8',  
                    '#7038F8',  
                   ]
```

```python
plt.figure(figsize=(11,6)) # Set plot dimensions
sns.swarmplot(x='Road Name', y='Car', data=database, 
              palette=road_key_colors)
```

```python
plt.figure(figsize=(11,6))
plt.ylim(-20,160) 
# Create plot
sns.violinplot(x='Road Name',
               y='Car', 
               data=database, 
               inner=None, # Remove the bars inside the violins
               palette=road_key_colors)

sns.swarmplot(x='Road Name',
               y='Car',  
              data=database, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Car Count by Road')
```
