```python
import pandas as pd
import os
os.chdir(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\\For Github\\")

#### CREATE LOOKUP INCLUDING DIRECTIONS BEFORE REDOING THIS
final_id_lookup = pd.read_csv(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\\For Github\\util\\Final_site_ID_lookup_with_directions.csv")
final_data = pd.read_csv(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\For Github\\5 - Factored_data\\2 - Data Factored to 2023\\Fully_factored_march_2023_flows.csv")

### MAKE SURE EQUIVALENT FIELD GIVEN IN LOOKUP
final_data["FINAL Site ID - Direction"] = final_data["FINAL Site ID"].astype(str) + " - " + final_data["direction"]
display(final_id_lookup)
```

```python
sites_not_included = final_id_lookup.loc[~final_id_lookup["FINAL Site ID - Direction"].isin(final_data["FINAL Site ID - Direction"])]
sites_included = final_id_lookup.loc[final_id_lookup["FINAL Site ID - Direction"].isin(final_data["FINAL Site ID - Direction"])]

```

```python
sites_not_included.to_csv(f"sites_not_included.csv", index = False)
```

```python
neutral_2019 = pd.read_csv(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\\For Github\\4 - Filtered_DBs\\1 - Neutral_Day_DBs\\1b - Neutral_Day_DBs_Converted&Appended_to_Hourly\\ATC_2019_Combined.csv")
neutral_2022 = pd.read_csv(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\\For Github\\4 - Filtered_DBs\\1 - Neutral_Day_DBs\\1b - Neutral_Day_DBs_Converted&Appended_to_Hourly\\ATC_2022_Combined.csv")
neutral_2023 = pd.read_csv(f"C:\\Users\\UKHRM004\\Documents\\HCC Model Update\\For Github\\4 - Filtered_DBs\\1 - Neutral_Day_DBs\\1b - Neutral_Day_DBs_Converted&Appended_to_Hourly\\ATC_2023_Combined.csv")

unfiltered_db = pd.concat([neutral_2019, neutral_2022, neutral_2023])

display(unfiltered_db)
```

```python
unfiltered_db["site - Source - direction"] = unfiltered_db["site"].astype(str) + " - " + unfiltered_db["Source"] + " - " + unfiltered_db["direction"]
sites_not_included["site - Source - direction"] = sites_not_included["site"].astype(str) + " - " + sites_not_included["Source"] + " - " + sites_not_included["direction"]
unfiltered_db_exc = pd.merge(unfiltered_db, sites_not_included[["site - Source - direction","FINAL Site ID", "FINAL Site ID - Direction", "Road", "District", "Town", "Road Name", "Road Type"]], on = "site - Source - direction", how = "inner")
display(unfiltered_db_exc)
```

```python
unfiltered_db_exc.to_csv(f"raw_data_excluded_from_final_sample.csv", index = False)
```

```python
no_days = lambda x: x.count() / 24.
rsd = lambda x: x.std() / x.mean()
unfiltered_db_exc_sample_size = unfiltered_db_exc
unfiltered_db_exc_sample_size["Number of Days"] = unfiltered_db_exc_sample_size.groupby(["FINAL Site ID", "FINAL Site ID - Direction"])["flow_total"].transform(no_days) #, "Road", "District", "Town", "Road Name", "Road Type", "direction"
unfiltered_db_exc_sample_size["RSD"] = unfiltered_db_exc_sample_size.groupby(["FINAL Site ID", "FINAL Site ID - Direction"])["flow_total"].transform(rsd) #, "Road", "District", "Town", "Road Name", "Road Type", "direction"

```

```python
display(unfiltered_db_exc_sample_size)
unfiltered_db_exc_sample_size.to_csv("TEST.csv")
```

```python
excluded_sites_table = unfiltered_db_exc_sample_size[["site", "Source", "direction", "FINAL Site ID", "Road", "District", "Town", "Road Name", "Road Type", "Number of Days", "RSD"]].groupby(["site", "Source", "direction", "FINAL Site ID"]).mean() #, "Road", "District", "Town", "Road Name", "Road Type"
excluded_sites_table.reset_index(inplace = True)
display(excluded_sites_table)
```

```python
excluded_sites_table.to_csv("excluded_sites_table.csv")
```
