```python
import numpy as np
import pandas as pd
import os
```

```python
#INPUTS
TIME_PERIODS = ["AM_peak_hr", "Inter_peak", "PM_peak_hr"]
USER_CLASSES = ["1_HBW", "2_HBEmp", "3_HBO", "4_NHBW", "5_NHBO", "6_LGV", "7_HGV"]
PROGRAM_DIRECTORY = os.getcwd()
FILE_DIRECTORY = f"{PROGRAM_DIRECTORY}\Finalising Outputs\Initial Outputs"
```

```python
new_rows = dict()
new_model_zones = [6030, 6140, 6141, 6148, 6200, 6201, 6202, 6203, 6204, 6205, 6206, 6207, 6208, 6210, 6211, 6500, 6501, 6502, 6503, 6504, 6505, 6506]

z = 0
for y in range(len(new_model_zones)):
    row = dict()
    for i in range(169):
        if i == 0:
            row[i] = new_model_zones[y]
        else:
            row[i] = 0
    new_rows[z] = row
    z += 1
```

```python
for time_period in TIME_PERIODS:
    matrices = dict()
    i = 0
    for user_class in USER_CLASSES:
        if user_class != "7_HGV":
            matrices[i] = pd.read_csv(f"{FILE_DIRECTORY}\{time_period}_{user_class}_Highway_Vehicle Trips_values.csv")
            matrices[i]= pd.DataFrame(np.vstack([matrices[i].columns, matrices[i]]))   
        else:
            matrices[i] = pd.read_csv(f"{FILE_DIRECTORY}\{time_period}_{user_class}_HGV_Vehicle Trips_values.csv")
            matrices[i]= pd.DataFrame(np.vstack([matrices[i].columns, matrices[i]]))   
        i += 1
    stacked_matrices = pd.DataFrame()
    for y in range(len(matrices)):
        for z in range(len(new_rows)):
            matrices[y] = matrices[y].append(new_rows[z], ignore_index=True)
        for z in range(len(new_model_zones)):
            matrices[y].insert((159+z), (10000+z), 0)
        matrices[y][0] = matrices[y][0].astype(int)
        matrices[y] = matrices[y].set_index(0).sort_index()
        stacked_matrices = pd.concat([stacked_matrices,matrices[y]], axis=0)
    stacked_matrices.to_csv(f"{PROGRAM_DIRECTORY}\Finalising Outputs\Final Outputs\{time_period}_Trips.csv", header=False)
    
    
            
```
